```python
def gpt(text, temp=0.4, max_t=32):
    try: 
        output = openai.Completion.create(
            model="text-davinci-003",
            prompt=text,
            temperature=temp,
            max_tokens=max_t,
            top_p=1.0,
            frequency_penalty=0.0,
            presence_penalty=0.0
        )
    except:
        print("Thinking...")
        time.sleep(1)
        return gpt(text, temp=0.4, max_t=32)
    return output
```

```python
zero_shot_prompt = lambda q, a: f"""Pregunta: {q}.
Respuesta de Felipe: {a}.
Responde con s o no 3 consultas distintas: 
Es correcta la respuesta dentro de la pregunta? 
Es correcta la respuesta de Felipe? 
Es coherente la respuesta de Felipe?
"""

q = "Antonella compr 23 dulces para su curso. Regal algunos dulces a sus compaeros y le quedan 9 en su bolsillo. Una amiga le dijo que regal 11 dulces en total. Es correcto lo que seala su amiga? Explica con tus palabras."
a = "no es verdad porque reste y me dio 14"
zero_shot_prompt(q, a)
```
