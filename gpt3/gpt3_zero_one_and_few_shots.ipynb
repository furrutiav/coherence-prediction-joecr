```python
def gpt(text, temp=0.4, max_t=32):
    try: 
        output = openai.Completion.create(
            model="text-davinci-003",
            prompt=text,
            temperature=temp,
            max_tokens=max_t,
            top_p=1.0,
            frequency_penalty=0.0,
            presence_penalty=0.0
        )
    except:
        print("Thinking...")
        time.sleep(1)
        return gpt(text, temp=0.4, max_t=32)
    return output
```

```python
zero_shot_prompt = lambda q, a: f"""Hola, ¿estas preparado? 
Eres un clasificador de incoherencia en respuestas a preguntas abiertas de alumnos de cuarto grado en matematicas. 
Te dare una pregunta y diferentes respuestas. 
Luego tienes que decir si la respuestas es coherente a la pregunta o bien incoherente a la pregunta.
Marcaremos las preguntas entre <P {{pregunta}} /P> y las respuestas por <R {{respuesta}} /R>.
Tu entregaras la etiqueta incoherente o bien coherente entre <E {{etiqueta}} /E>.
¿Estas listo?
<P {q} /P>. 
<R {a} /R>.
<E """

q = "Antonella compró 23 dulces para su curso. Regaló algunos dulces a sus compañeros y le quedan 9 en su bolsillo. Una amiga le dijo que regaló 11 dulces en total.¿ Es correcto lo que señala su amiga? Explica con tus palabras."
a = "no es verdad porque reste y me dio 14"
zero_shot_prompt(q, a)
```

```python
%%time
responses = []
for i, ix in enumerate(test.index):
    q, a = test.loc[ix]["Q A".split()]
    o = gpt(zero_shot_prompt(str(q), str(a)), temp=0.4, max_t=10)["choices"][0]["text"]
    responses.append([ix, o])
    print(100*(i+1)/test.shape[0], ix, o)
```

```python
test_v1 = test.copy()
test_v1["gpt"] = [x[1] for x in responses]
test_v1["gpt-label"] = [1 if "incoherente" in x[1].lower() else 0 for x in responses]
test_v1
```
